we are going to use tfdif , becuase some words are important too, so give some weights to them.Data preprocessing also need to be done

```python
import re
```

```python
def wordpot(text):
    text = text.lower()

    # remove URLs
    text = re.sub(r"https?://\S+|www\.\S+", "", text)

    #remove HTML tags
    text = re.sub(r"<.*?>", "", text)

    #remove punctuations
    text = re.sub(r"[^\w\s]", "", text)

    #remove digits 
    text = re.sub(r"\d+", "", text)

    #remove newline characters
    text = re.sub(r"\n", "", text)
  
    return text
```

```python
news['text'] = news['text'].apply(wordpot)
news.head()
```

```python
x = news['text']
y= news['label']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorization = TfidfVectorizer()
```

```python
xv_train =vectorization.fit_transform(x_train)
```

```python
xv_test = vectorization.transform(x_test)
```

```python
pred_lr = LR.predict(xv_test)
pred_lr
```

```python
LR.score = LR.score(xv_test, y_test)
LR.score
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, pred_lr))
```

```python
def output_label(n):
    if n == 0:
        return "It is a Fake News"
    else:  
        return "It is a genuine News"
```

```python
def manual_testing(news):
    testing_news = {"text": [news]}
    new_df_test = pd.DataFrame(testing_news)
    new_df_test['text'] = new_df_test['text'].apply(wordpot)
    new_x_test = new_df_test['text']
    new_xv_test = vectorization.transform(new_x_test)
    pred_lr = LR.predict(new_xv_test)
    # pred_gbc = gbc.predict(new_xv_test)
    pred_dtc = DTC.predict(new_xv_test)
    pred_rfc = rfc.predict(new_xv_test)
    return (
        f"\n\n LR Predictions :{output_label(pred_lr[0])}"
        
        f"\n\n DTC Predictions :{output_label(pred_dtc[0])}"
        f"\n\n RFC Predictions :{output_label(pred_rfc[0])}"
    )
```

```python
news_article = str(input())
```

```python
manual_testing(news_article)
```
